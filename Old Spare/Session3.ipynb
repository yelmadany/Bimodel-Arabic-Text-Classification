```python
#Separating features X from labels y
y = df['HeartDisease'].to_numpy()
X = df.drop(columns=['HeartDisease']).to_numpy()
```

```python
#print X y
```

```python
#Next Scaling
from sklearn.preprocessing import MinMaxScaler #import
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
X
```

```python
#Divide dataset into training and testing
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
#ML
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(max_depth = 4)
clf.fit(X_train,y_train)
y_predict = clf.predict(X_test)
```

```python
y_predict #predicted label
```

```python
y_test #true labels
```

```python
clf.score(X_test, y_test) #testing accuracy
```

```python
clf.score(X_train, y_train) #training accuracy
```

```python
from sklearn.metrics import f1_score, confusion_matrix, accuracy_score, precision_recall_fscore_support, classification_report

f1_score(y_test, y_predict)
```

```python
confusion_matrix(y_test, y_predict)
```

```python
accuracy_score(y_test, y_predict) #similar to clf.score
```

```python
precision_recall_fscore_support(y_test, y_predict)
```

```python
print(classification_report(y_test, y_predict))
```

```python
clf.tree_
```

```python
from sklearn import tree

tree.plot_tree(clf)
```

```python
from sklearn.model_selection import cross_validate

results = cross_validate(clf, X, y, cv=4, return_train_score =  True, scoring=('accuracy','f1','recall','precision','roc_auc'))
results
```
