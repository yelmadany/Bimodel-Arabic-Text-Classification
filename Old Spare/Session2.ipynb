```python
#Separating features X from labels y
y = df['HeartDisease'].to_numpy()
X = df.drop(columns=['HeartDisease']).to_numpy()
```

```python
#print X y
```

```python
#Next Scaling
from sklearn.preprocessing import MinMaxScaler #import
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
X
```

```python
#Divide dataset into training and testing
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
#ML
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(max_depth = 4)
clf.fit(X_train,y_train)
y_predict = clf.predict(X_test)
```

```python
y_predict #predicted label
```

```python
y_test #true labels
```

```python
clf.score(X_test, y_test) #testing accuract
```

```python
clf.score(X_train, y_train) #training accuracy
```
