```python
import pandas as pd

df = pd.read_csv('Final Data (1).csv')
```

```python
print(df.shape)
```

```python
df.head(20)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.to_csv('Final Data_Cleaned.csv')
```

```python
y = df['Gender'].to_numpy()
X = df.drop(columns=['Gender']).to_numpy()
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import MinMaxScaler
filenames = X[:, 0]
numeric_data = X[:, 1:]
scaler = MinMaxScaler()
X = scaler.fit_transform(numeric_data)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(max_depth = 8)
clf.fit(X_train,y_train)
y_predict = clf.predict(X_test)
```

```python
y_predict
```

```python
y_test
```

```python
clf.score(X_test, y_test)
```

```python
clf.score(X_train, y_train)
```

```python
from sklearn.metrics import f1_score, confusion_matrix, accuracy_score, precision_recall_fscore_support, classification_report

f1_score(y_test, y_predict)
```

```python
confusion_matrix(y_test, y_predict)
```

```python
accuracy_score(y_test, y_predict)
```

```python
precision_recall_fscore_support(y_test, y_predict)
```

```python
print(classification_report(y_test, y_predict))
```

```python
from sklearn.model_selection import cross_validate

results = cross_validate(clf, X, y, cv=4, return_train_score =  True, scoring=('accuracy','f1','recall','precision','roc_auc'))
results
```
