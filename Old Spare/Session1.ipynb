```python
from sklearn.preprocessing import OrdinalEncoder #import
ord_enc = OrdinalEncoder()  #create object

df[['Sex', 'ExerciseAngina', 'ST_Slope']] = ord_enc.fit_transform(df[['Sex', 'ExerciseAngina', 'ST_Slope']] )
#Exercise: Apply ordinal encoding in order
```

```python
from sklearn.preprocessing import OneHotEncoder #import
onehot_enc = OneHotEncoder()  #create object

onehot_enc_result = onehot_enc.fit_transform(df[['ChestPainType']])
onehot_enc_result.toarray()
```

```python
onehot_enc.categories_
```

```python
new_df = pd.DataFrame(onehot_enc_result.toarray(), columns = onehot_enc.categories_)
df = df.join(new_df)
```
